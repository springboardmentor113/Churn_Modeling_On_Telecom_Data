5.Outlier Treatment :

a. Boxplot : Q3+(1.5*IQR)&Q1-(1.5*IQR)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Add this import statement

# Assuming 'df' is your DataFrame
#df = pd.read_csv('churn_data.csv')

# Define a function to remove outliers using the IQR method
def remove_outlier_iqr(col):
    sorted_col = sorted(col)
    q1, q3 = np.percentile(sorted_col, [25, 75])
    IQR = q3 - q1
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    return lower_bound, upper_bound

# Define a function to remove outliers using quartiles
def remove_outlier_quantile(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    low = q1 - 1.5 * iqr
    high = q3 + 1.5 * iqr
    return low, high

# Iterate over each column
for col in df.columns:
    # Choose one method for outlier removal based on your preference
    # For example, using the IQR method:
    lower_bound, upper_bound = remove_outlier_iqr(df[col])
    
    # Apply outlier removal to the column
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])

# Boxplot to visualize the distribution after outlier removal for all columns
df.boxplot(figsize=(10,6))
plt.title("Boxplot of all Columns after Outlier Removal")
plt.xticks(rotation=90)
plt.show()
df.shape
```

b. standardisation: +/-3 sigma 

```python
# standadization: +/-3 sigma approach

def standardize_data(df):
  # Calculate mean and standard deviation of the data
  mean = df.mean(axis=0)
  std = df.std(axis=0)
  # Standardize the data
  standardized_df = (df - mean) / std

  # Return the standardized data
  return standardized_df

# Standardize the data
standardized_df = standardize_data(df)

# Print the standardized data
print("standard\n",standardized_df)
```

```python
# prompt: capping and flooring

# Capping and flooring
def cap_floor_data(df, cap, floor):
  # Cap the data
  capped_df = df.clip(upper=cap)
  # Floor the data
  floored_df = capped_df.clip(lower=floor)
  # Return the capped and floored data
  return floored_df

# Cap and floor the data
capped_floored_df = cap_floor_data(df, cap=4, floor=-4)

# Print the capped and floored data
print("Capped and Floored Data\n", capped_floored_df)
```

c. Impute with  Mean/Median ,if variable  is numeric and with Mode  if  variable  is categorical

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Identify numeric and categorical columns
numeric_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(exclude=np.number).columns

# Impute missing values in numeric columns with the median
imputer_numeric = SimpleImputer(missing_values=np.nan, strategy='median')
df[numeric_cols] = imputer_numeric.fit_transform(df[numeric_cols])

# Impute missing values in categorical columns with the mode
if not categorical_cols.empty:
    imputer_categorical = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    df[categorical_cols] = imputer_categorical.fit_transform(df[categorical_cols].values)

# Print the updated DataFrame
print(df)
```

```python

import pandas as pd

# Calculate the correlation matrix
correlation_matrix = df.corr()


# Select the columns with high correlation
highly_correlated_cols = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            highly_correlated_cols.append(correlation_matrix.columns[j])

# Drop the highly correlated columns
df = df.drop(highly_correlated_cols, axis=1)

# Print the updated DataFrame
print(df)
```
