```python
def handle_outliers(df):
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
            df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    
    return df

df = handle_outliers(df)
```

```python
print(df.corr())
```

```python
corr_mat = df.corr().abs()
cols_to_remove = set()
threshold = 0.8
for col in corr_mat.columns:
    if col in cols_to_remove:
        continue
    correlated_cols = corr_mat[col][(corr_mat[col] > threshold) & (corr_mat[col].index != col)].index.tolist()
    cols_to_remove.update(correlated_cols)
if cols_to_remove:
    df.drop(columns=cols_to_remove, inplace=True)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
X = df.drop('target',axis=1)
vif["Feature"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
high_vif = vif[vif['VIF'] > 20]
print(len(high_vif))
print(high_vif.sort_values(by = 'VIF', ascending = False))
```

```python
df = df.drop(high_vif['Feature'],axis = 1)
```

```python
df=pd.read_csv("modified_dataRushikesh.csv")
df.head
```

```python
df.shape
```

```python
df['target'] = df['target'].astype('int64')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#spliting of Data into Train ,Test
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
#Making of Seperate files for training and the testing
train_df.to_csv('training_data.csv', index=False)
test_df.to_csv('testing_data.csv', index=False)
```

```python
X_train = train_df.drop('target', axis=1) 
Y_train = train_df['target']
```

```python
X_test = test_df.drop('target', axis=1)
Y_test = test_df['target']
```

```python
#Random Forest Model
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier
```

```python
#Fitting of data in the random Forest
rf_classifier.fit(X_train, Y_train)
```

```python
Y_pred = rf_classifier.predict(X_test)
#accuracy code                                        
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy on test data: {accuracy:}')
```

```python
#Presion SCore For the model
from sklearn.metrics import precision_score
precision = precision_score(Y_test, Y_pred)
print(f'Precision of test data:{precision}')
```

```python
# Confusion Matrix 
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_pred)
cm_df = pd.DataFrame(cm, columns=['Predicted Negative', 'Predicted Positive'], 
                     index=['Actual Negative', 'Actual Positive'])
print(f'Confusion_matrix:{cm_df}')
```

```python
#Recall score
from sklearn.metrics import recall_score

recall = recall_score(Y_test, Y_pred)
print(f'Recall_Score:{recall}')
```

```python
#f1_Score
from sklearn.metrics import f1_score

f1 = f1_score(Y_test, Y_pred)
print(f'f1-score:{f1}')
```
