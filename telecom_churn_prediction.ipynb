### 7. removing the highly correlated variables


```python
correlation_matrix = df.corr().abs()
correlation_matrix
```

```python
def find_highly_correlated_features(correlation_matrix, threshold=0.9, target_col="target"):
    highly_correlated_features = {}
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if i != correlation_matrix.columns.get_loc(target_col) and j != correlation_matrix.columns.get_loc(target_col) and correlation_matrix.iloc[i, j] > threshold:
                var_i = correlation_matrix.columns[i]
                var_j = correlation_matrix.columns[j]
                correlation_value = correlation_matrix.iloc[i, j]
                highly_correlated_features[(var_i, var_j)] = correlation_value

    return highly_correlated_features

target_col = "target"  
threshold = 0.9

highly_correlated_pairs = find_highly_correlated_features(correlation_matrix, threshold, target_col)
print("Pairs of highly correlated variables (excluding target) with their correlation values:")

    
for pair, correlation_value in sorted(highly_correlated_pairs.items(), key=lambda x: x[1]):
    print(pair, ":", correlation_value)

for var_i, var_j in highly_correlated_pairs:
    if var_i in df.columns:
        df.drop(columns=[var_i], inplace=True)

print("DataFrame after removing highly correlated variables:")
print(df.head())
```

### 8. multicollinearity (VIF >5)

```python
def calculate_vif(X):
    corr_matrix = np.corrcoef(X, rowvar=False)
    inv_corr_matrix = np.linalg.inv(corr_matrix)
    vif_values = np.diagonal(inv_corr_matrix)
    return vif_values

vif_threshold = 5

vif_values = calculate_vif(df)

vif_results = pd.DataFrame({'Variable': df.columns, 'VIF': vif_values})

high_vif_variables = vif_results[vif_results['VIF'] > vif_threshold]

if not high_vif_variables.empty:
    print("Variables with VIF greater than", vif_threshold)
    print(high_vif_variables)
else:
    print("No variables with VIF greater than", vif_threshold)
```

### Logistic Regressor

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


# Splitting the data into training and testing sets
X = df.drop(columns=['target']) 
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

model = LogisticRegression(max_iter=10)

model.fit(X_train, y_train)


y_pred_train = model.predict(X_train)
accuracy_train = accuracy_score(y_train, y_pred_train)
print("Training Accuracy:", accuracy_train)
print("Classification Report (Training Data):")
print(classification_report(y_train, y_pred_train))

y_pred_test = model.predict(X_test)
accuracy_test = accuracy_score(y_test, y_pred_test)
print("Test Accuracy:", accuracy_test)
print("Classification Report (Test Data):")
print(classification_report(y_test, y_pred_test))
```

### hyper-parameter tuning

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Assuming X and y are defined and the data is already loaded in df
# X = df.drop(columns=['target']) 
# y = df['target']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize logistic regression model with increased max_iter
log_reg = LogisticRegression(max_iter=1000)

# Define the parameter grid with compatible solver-penalty combinations
param_grid = [
    {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'penalty': ['l1'], 'solver': ['liblinear', 'saga']},
    {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'penalty': ['l2'], 'solver': ['liblinear', 'saga', 'newton-cg', 'lbfgs']}
]

# Perform grid search cross-validation
grid_search = GridSearchCV(estimator=log_reg, param_grid=param_grid, cv=5, scoring='accuracy', error_score='raise')
grid_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_


# Evaluate the best model on the training set
y_pred_train_best = best_model.predict(X_train_scaled)
accuracy_train_best = accuracy_score(y_train, y_pred_train_best)
print("Train Accuracy (Best Model):", accuracy_train_best)
print("Classification Report (Best Model - Train Data):")
print(classification_report(y_train, y_pred_train_best))


# Evaluate the best model on the test set
y_pred_test_best = best_model.predict(X_test_scaled)
accuracy_test_best = accuracy_score(y_test, y_pred_test_best)
print("Test Accuracy (Best Model):", accuracy_test_best)
print("Classification Report (Best Model - Test Data):")
print(classification_report(y_test, y_pred_test_best))
```
