```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Generate sample data (replace with your dataset)
np.random.seed(0)
X = np.random.normal(loc=0, scale=1, size=(100, 2))
X[0, 0] = 10  # introduce outlier

# Step 1: Identify outliers using z-score
z_scores = np.abs((X - X.mean()) / X.std())
threshold = 3
outliers = np.where(z_scores > threshold)

# Step 2: Remove outliers
X_clean = np.delete(X, outliers[0], axis=0)

# Step 3: Standardize the cleaned data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)

print("Original data shape:", X.shape)
print("Cleaned data shape:", X_clean.shape)
print("Scaled data mean:", X_scaled.mean(axis=0))
print("Scaled data std:", X_scaled.std(axis=0))
```

```python
def remove_outliers(df, columns):
    for column in columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
columns_for_outlier_removal = ['s1.new.rev.m1', 's4.usg.ins.p2']  
df_cleaned = remove_outliers(df, columns_for_outlier_removal)
df_cleaned.to_excel('fin_churn.xlsx', index=False) 
df_cleaned
```

# Removing outliers using Capping & Flooring

```python
import pandas as pd
df = pd.read_excel('fin_churn.xlsx')  
def remove_outliers_cap_floor(df, columns, lower_percentile=0.05, upper_percentile=0.95):
    treated_df = df.copy()
    for column in columns:
        lower_bound = df[column].quantile(lower_percentile)
        upper_bound = df[column].quantile(upper_percentile)
        treated_df[column] = treated_df[column].clip(lower=lower_bound, upper=upper_bound)
    return treated_df
columns_for_outlier_treatment = ['s4.usg.ins.p2','s2.rch.val.p6'] 
df_treated = remove_outliers_cap_floor(df, columns_for_outlier_treatment)
df_treated.to_excel('fin_churn.xlsx', index=False) 
df_treated
```

# 7.Removing highly correlated variables

```python
correlation_matrix = df.corr().abs()
threshold = 0.8  
highly_correlated_pairs = np.where(correlation_matrix > threshold)
columns_to_remove = set()
for i, j in zip(*highly_correlated_pairs):
    if i != j and i not in columns_to_remove and j not in columns_to_remove:
        col_i = correlation_matrix.columns[i]
        col_j = correlation_matrix.columns[j]
        if col_i not in columns_to_remove:
            columns_to_remove.add(col_i)
df_cleaned = df.drop(columns=columns_to_remove)
df_cleaned.to_excel('fin_churn.xlsx', index=False) 
df_cleaned
```

# 8.Multicollinearity(VIF>5)

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
df = pd.read_excel('fin_churn.xlsx')  
def calculate_vif(df, thresh=5.0):
    variables = list(range(df.shape[1]))
    dropped = True
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(df.iloc[:, variables].values, ix) for ix in range(df.iloc[:, variables].shape[1])]
        max_vif = max(vif)
        if max_vif > thresh:
            maxloc = vif.index(max_vif)
            print(f'Dropping {df.iloc[:, variables].columns[maxloc]} with VIF={max_vif}')
            del variables[maxloc]
            dropped = True
    print('Remaining variables:')
    print(df.columns[variables])
    return df.iloc[:, variables]
df_no_multicollinearity = calculate_vif(df)
df_no_multicollinearity.to_excel('fin_churn.xlsx', index=False) 
df_no_multicollinearity
```

# Performing Hyperparameter tuning on the model

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
data = df_no_multicollinearity
X = data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_search = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
y_pred = best_model.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
