# Removing outliers in the dataset using boxplot technique

```python
import pandas as pd
import matplotlib.pyplot as plt
# Define a function to remove outliers from columns
def remove_outlier_columns(df):
    outlier_columns = []
    for column in df.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        if not outliers.empty:
            outlier_columns.append(column)
    df_cleaned = df.drop(outlier_columns, axis=1)
    return df_cleaned
data.boxplot()
for column in data.columns:
    data = remove_outlier_columns(data)
data.boxplot()
plt.show()
```

# Removing HighlyCorrelated values

```python
import pandas as pd
import numpy as np
correlation_matrix = data.corr().abs()
highly_correlated_mask = correlation_matrix > 0.95
np.fill_diagonal(highly_correlated_mask.values, False)
correlated_pairs = np.where(highly_correlated_mask)
features_to_remove = set()
for feature1, feature2 in zip(*correlated_pairs):
    if feature1 < feature2:
        features_to_remove.add(feature1)
    else:
        features_to_remove.add(feature2)

data = data.drop(data.columns[list(features_to_remove)], axis=1)

```
