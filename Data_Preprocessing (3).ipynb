```python
# Using dataframe df: box plot maximize the plot so we can see the data pounts crealy

import seaborn as sns
import matplotlib.pyplot as plt
# Calculate Q1, Q3, and IQR
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

# Define lower and upper bounds for outlier detection
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

plt.figure(figsize=(30, 10))
sns.boxplot(data=df)
plt.show()
df.shape
```

```python
def remove_outliers_boxplot(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return column[(column >= lower_bound) & (column <= upper_bound)]

# Iterate over each column
for col in df.columns:
    # Remove outliers from the current column using the boxplot method
    df[col] = remove_outliers_boxplot(df[col])
plt.figure(figsize=(30, 10))
sns.boxplot(df)
df.shape
```

```python
# standadization: +/-3 sigma approach

def standardize_data(df):
  # Calculate mean and standard deviation of the data
  mean = df.mean(axis=0)
  std = df.std(axis=0)
  # Standardize the data
  standardized_df = (df - mean) / std

  # Return the standardized data
  return standardized_df

# Standardize the data
standardized_df = standardize_data(df)

# Print the standardized data
print("standard\n",standardized_df)
```

```python
#capping and flooring

# Capping and flooring
def cap_floor_data(df, cap, floor):
  # Cap the data
  capped_df = df.clip(upper=cap)
  # Floor the data
  floored_df = capped_df.clip(lower=floor)
  # Return the capped and floored data
  return floored_df

# Cap and floor the data
capped_floored_df = cap_floor_data(df, cap=4, floor=-4)

# Print the capped and floored data
print("Capped and Floored Data\n", capped_floored_df)
```

```python
#missing value tratment:remove record s if NA are less than 5%

# Calculate the percentage of missing values per column
missing_values_percentage = df.isna().sum().div(df.shape[0]) * 100

# Select columns where the percentage of missing values is less than 5%
columns_to_keep = missing_values_percentage[missing_values_percentage < 5].index

# Drop rows with missing values in the selected columns
df = df[columns_to_keep].dropna()

# Print the updated DataFrame
print(df)
```

```python
#missing value tratment:remove if NA are 50% in any variable

df = df.dropna(thresh=0.5)
df
```

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Identify numeric and categorical columns
numeric_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(exclude=np.number).columns

# Impute missing values in numeric columns with the median
imputer_numeric = SimpleImputer(missing_values=np.nan, strategy='median')
df[numeric_cols] = imputer_numeric.fit_transform(df[numeric_cols])

# Impute missing values in categorical columns with the mode
if not categorical_cols.empty:
    imputer_categorical = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    df[categorical_cols] = imputer_categorical.fit_transform(df[categorical_cols].values)

# Print the updated DataFrame
print(df)
```

```python
# removing highly correlated variales

import pandas as pd

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Select the columns with high correlation
highly_correlated_cols = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            highly_correlated_cols.append(correlation_matrix.columns[j])

# Drop the highly correlated columns
df = df.drop(highly_correlated_cols, axis=1)

# Print the updated DataFrame
print(df)
```

```python
# multicollinearity (VIF>5)

import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

# Calculate VIF for each variable
vif_df = pd.DataFrame()
vif_df["features"] = df.columns
vif_df["VIF Factor"] = [vif(df.values, i) for i in range(df.shape[1])]

# Print VIF for each variable
print(vif_df)

# Identify variables with VIF > 5
high_vif_cols = vif_df[vif_df['VIF Factor'] > 5]['features'].tolist()

# Drop variables with high VIF
df = df.drop(high_vif_cols, axis=1)

# Print the updated DataFrame
print(df)
```

```python
# Importing the necessary library
!pip install scikit-learn
from sklearn.model_selection import train_test_split

# Assuming X and y are already defined as your feature set and target variable respectively
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=3, min_samples_leaf=5)

# Fit the model with training data
classifier.fit(X_train, y_train)
```

```python
#Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Calculating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the Decision Tree Classifier:", accuracy)
```

```python

# Calculate probabilities for the entire dataset
class_labels_y, counts_y = np.unique(y, return_counts=True)
probabilities_y = counts_y / counts_y.sum()

# Calculate the entropy for the entire dataset
entropy_dataset = -np.sum(probabilities_y * np.log2(probabilities_y))
print("Entropy of the dataset:", entropy_dataset)
```

```python
# Calculate probabilities for the test set
class_labels_y_test, counts_y_test = np.unique(y_test, return_counts=True)
probabilities_y_test = counts_y_test / counts_y_test.sum()

# Calculate the entropy for the test set
entropy_test_set = -np.sum(probabilities_y_test * np.log2(probabilities_y_test))
print("Entropy of the test set:", entropy_test_set)
```
